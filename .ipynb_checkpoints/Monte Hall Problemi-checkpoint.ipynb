# 100 deneme için bakalım

```python
import random
```

```python
epoch = 100
```

```python
score_one_choice = 0
score_two_choice = 0
for i in range(epoch):
    kapilar = [1,2,3]
    oduller = ['araba', 'keci_1', 'keci_2']
    odul_eslesme = {}
    
    
    for j in kapilar:
        choice = random.choice(oduller)
        oduller.remove(choice)
        odul_eslesme[j] = choice
        
    # One Choice    
    final_choice_one = random.choice(kapilar)
    if odul_eslesme[final_choice_one] == 'araba':
        score_one_choice += 1
        
    # Two Choice
    two_choice_1 = random.choice(kapilar)
    
    if odul_eslesme[two_choice_1] == 'araba':
        kapilar.remove(two_choice_1)
        secimler_2 = random.choice(kapilar)
        kapilar.remove(secimler_2)
        secim_3 = kapilar[0]
        if odul_eslesme[secim_3] == 'araba':
            score_two_choice += 1            
        
    elif odul_eslesme[two_choice_1] == 'keci_1':
        for kapi,odul in odul_eslesme.items():
            if odul == 'keci_2':
                ind = kapi
        kapilar.remove(two_choice_1)
        kapilar.remove(ind)
        secim_3 = kapilar[0]
        if odul_eslesme[secim_3] == 'araba':
            score_two_choice += 1
        
    elif odul_eslesme[two_choice_1] == 'keci_2':
        for kapi,odul in odul_eslesme.items():
            if odul == 'keci_1':
                ind = kapi
        kapilar.remove(two_choice_1)
        kapilar.remove(ind)
        secim_3 = kapilar[0]
        if odul_eslesme[secim_3] == 'araba':
            score_two_choice += 1
            
            
print("One Choice Score : ", score_one_choice)
print("Two Choice Score : ", score_two_choice)
        
```

# 100 tane 100 deneme yapıp bakarsak

```python
scores_1_ch = []
scores_2_ch = []
for i in range(100):
    score_one_choice = 0
    score_two_choice = 0
    for i in range(epoch):
        kapilar = [1,2,3]
        oduller = ['araba', 'keci_1', 'keci_2']
        odul_eslesme = {}


        for j in kapilar:
            choice = random.choice(oduller)
            oduller.remove(choice)
            odul_eslesme[j] = choice

        # One Choice    
        final_choice_one = random.choice(kapilar)
        if odul_eslesme[final_choice_one] == 'araba':
            score_one_choice += 1

        # Two Choice
        two_choice_1 = random.choice(kapilar)

        if odul_eslesme[two_choice_1] == 'araba':
            kapilar.remove(two_choice_1)
            secimler_2 = random.choice(kapilar)
            kapilar.remove(secimler_2)
            secim_3 = kapilar[0]
            if odul_eslesme[secim_3] == 'araba':
                score_two_choice += 1            

        elif odul_eslesme[two_choice_1] == 'keci_1':
            for kapi,odul in odul_eslesme.items():
                if odul == 'keci_2':
                    ind = kapi
            kapilar.remove(two_choice_1)
            kapilar.remove(ind)
            secim_3 = kapilar[0]
            if odul_eslesme[secim_3] == 'araba':
                score_two_choice += 1

        elif odul_eslesme[two_choice_1] == 'keci_2':
            for kapi,odul in odul_eslesme.items():
                if odul == 'keci_1':
                    ind = kapi
            kapilar.remove(two_choice_1)
            kapilar.remove(ind)
            secim_3 = kapilar[0]
            if odul_eslesme[secim_3] == 'araba':
                score_two_choice += 1

    scores_1_ch.append(score_one_choice)
    scores_2_ch.append(score_two_choice)

print("1 ch : ",scores_1_ch)
print("2 ch : ",scores_2_ch)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,9))
plt.plot(range(100), scores_1_ch,label = 'Tek Secim')
plt.plot(range(100), scores_2_ch, label = 'İki Secim', color='r')
plt.legend()
plt.show()
```

```python
print("Tek Seçim Ortalaması : ",sum(scores_1_ch)/100)
print("İki Seçim Ortalaması : ",sum(scores_2_ch)/100)
```
